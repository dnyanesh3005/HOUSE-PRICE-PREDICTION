```python
df1=df.drop(['id','date','waterfront','view','yr_renovated','sqft_above','sqft_basement','yr_built','zipcode','lat','long','sqft_living15','sqft_lot15'],axis=1)
```

```python
df1.drop_duplicates()
```

```python
df1=df1.astype(int)
```

```python
df1.dtypes
```

```python
x=df1.iloc[:,1:5].head(1000)
y=df1['price'].head(1000)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=2)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_test
```

```python
r2=r2_score(y_test,y_pred)
```
